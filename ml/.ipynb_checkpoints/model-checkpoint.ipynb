## Required Libraries

```python
import pandas as pd 
import numpy as np 
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

## Importing/Cleaning the Data 

```python
df = pd.read_csv("../data/creditcard.csv")

df
```

```python
# dimensions of the data 
print(f"Data Dimensions: {df.shape}")

#Duplicate Rows
print(f"Duplicated entries: {df.duplicated().sum()}")

# Check missing values
print(df.isnull().sum().sort_values(ascending=False).head(20))
```

## Prepping the Data for XGBoost

```python
from sklearn.preprocessing import OneHotEncoder 
from sklearn.compose import ColumnTransformer

x = df.drop(columns=["is_fraud"])

y = df["is_fraud"]

categorical_cols = ["time_of_day", "merchant_category", "device_type", "country"]

#Using OneHotEncoder for categorical data 
encoder = OneHotEncoder(drop='first', sparse_output=False)

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', encoder, categorical_cols)
    ],
    remainder='passthrough' # Keep all other numeric columns (amount, account_age, etc.) untouched
)

x_encoded = preprocessor.fit_transform(x) 

encoded_feature_names = preprocessor.get_feature_names_out()

df_cleaned = pd.DataFrame(x_encoded, columns=encoded_feature_names)

df_cleaned["is_fraud"] = y.values

df_cleaned
```

```python
#Save the dataset
df_cleaned.to_csv("../data/cleaned_credit_card.csv")
```

## Splitting the data and labels 

```python
## Prepping the data for training 

x = df_cleaned.drop(columns=["is_fraud"])

y = df_cleaned["is_fraud"]
```

```python

x_train, x_test, y_train, y_test = train_test_split(
    x, y, 
    test_size=0.2,
    random_state=42
)
```

## Train the XGBoost Model

```python
from xgboost import XGBClassifier

model = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric="logloss"
)

# STEP 1: train
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report

probs = model.predict_proba(x_test)[:, 1]

y_pred = (probs > 0.4).astype(int)

print(classification_report(y_test, y_pred))
```

## Save the Model/Features into AWS S3 Bucket

```python
import boto3
import joblib
import io

s3 = boto3.client("s3")

BUCKET = "sentinelflow-models-123"


# Save model into memory buffer
buffer1 = io.BytesIO()
joblib.dump(model, buffer1)
buffer1.seek(0)


# Save feature columns
buffer2 = io.BytesIO()
joblib.dump(x_train.columns, buffer2)
buffer2.seek(0)


s3.upload_fileobj(
    buffer1,
    BUCKET,
    "models/sentinelflow_xgb_v1.pkl"
)

s3.upload_fileobj(
    buffer2,
    BUCKET,
    "models/feature_columns_v1.pkl"
)

print("Model + features uploaded directly to S3 successfully")
```

## Evalauting other metrics (Confusion Matrix, etc.) 

```python
 from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

print(cm)
```
